```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'F:\GUVI_DATA_SCIENCE\Project\Hate_Speech_Classification\Hate_Speech_Dataset\final_processed.csv')
df.head()
```

```python
hate_df = df.query("label == 'hate'")
```

```python
hate_df.isnull().sum()
```

```python
hate_df = hate_df.dropna()
```

```python
text_data = ' '.join(hate_df['text'])
text_data
```

```python
wc = WordCloud(background_color='white',
               height = 600,
               width = 600)
```

```python
wc.generate(text_data)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(wc)
```
